```python
import pandas as pd

# create sample Data

data = {
    'Month': ['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun'],
    'Product A': [150,200,180,220,210,260],
    'Product B': [100,120,130,140,160,180],
    'Product C': [90,110,115,100,130,150]
 }

df = pd.DataFrame(data)

print(df)
df.to_excel("sampleData.xlsx")
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Month'], df['Product A'], marker='o', label='Product A')
plt.plot(df['Month'], df['Product B'], marker='o', label='Product B')
plt.plot(df['Month'], df['Product C'], marker='o', label='Product C')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()


plt.savefig('monthly_sales_trend.png')
plt.show()
```

```python
june_sales = df[df['Month'] == 'Jun'][['Product A', 'Product B', 'Product C']].values.flatten()

plt.bar(['Product A', 'Product B', 'Product C'], june_sales, color=['blue', 'green', 'orange'])
plt.title('June Sales Comparison')
plt.ylabel('Sales')
plt.show()
```

```python
labels=['Product A', 'Product B', 'Product C']
april_sales = df[df['Month'] == 'Apr'][labels].values.flatten()


plt.pie(april_sales, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Product Sales Share in April')
plt.axis('Equal')
plt.show()
```

```python
labels=['Product A', 'Product B', 'Product C']
april_sales = df[df['Month'] == 'Apr'][labels].values.flatten()


plt.pie(april_sales, labels=labels, autopct='%1.1f%%', startangle=140, pctdistance=0.8,labeldistance=0.95)
plt.title('Product Sales Share in April')
plt.axis('Equal')
plt.show()
```

```python
# load sample dataset (inside seaborn module)
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
df = sns.load_dataset('tips') #restaurant bills
df
```

```python
sns.barplot(data=df,x='day',y='total_bill', estimator='mean')
plt.title('Average Bill per Day')
plt.show()
```

```python
sns.boxplot(data=df,x='day',y='total_bill', hue='sex')
plt.title('Bill Distrubution by Day and Gender')
plt.show()
```

```python
sns.boxplot(data=df,x='day',y='total_bill', hue='sex',palette={'Male':'skyblue','Female':'red'})
plt.title('Bill Distrubution by Day and Gender')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', s=200)
plt.show
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', s=200)
sns.lmplot(data=df, x='total_bill', y='tip', hue='sex', scatter_kws={'s':200})
plt.show
```
