```python
#lineplot
plt.plot(x,y)
plt.title('Lineplot of x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show
```

```python
#lineplot
from random import randint

def randCol():
    
    return f"#{randint(0, 0xFFFFFF):06x}"
randCol()
# colorhex = str(hex(randint(0,16777215)))
# hex(randint(0,16777215))



plt.plot(x,y, color=randCol(), linestyle= '--')
plt.title('Lineplot of x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show
```

```python
z = [5] * len(x)
plt.subplot(2,1,2)
plt.plot(x,y, color=randCol(),linestyle='--', linewidth=3)
plt.title('lineplot between x and y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(2,1,1)
plt.plot(x,z, color=randCol(),linestyle=':', linewidth=2)
plt.title('lineplot between x and z')
plt.xlabel('x axis')
plt.ylabel('z axis')
plt.grid(True)
plt.subplots_adjust(hspace=0.5) #or plt.tight_layout()
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()
 
```

```python
# ---- imports ---------------------------------------------------------------
from pathlib import Path
import matplotlib.pyplot as plt
from moviepy import ImageSequenceClip   # MoviePy  2.x  (no .editor module)

# ---- static data -----------------------------------------------------------
fruit     = ['Apple', 'Orange', 'Grape', 'Dragon']
quantity  = [23, 55, 23, 85]

# ---- frame generation ------------------------------------------------------
N      = 120                 # total frames
frames = Path("./datasets/chartfigs/chartvids/")
frames.mkdir(exist_ok=True)

for i in range(N):
    plt.figure(figsize=(4, 4))
    plt.pie(
        quantity,
        labels      = fruit,
        autopct     = '%0.1f%%',
        colors      = [randCol() for _ in fruit],  # 4 random colours / frame
        startangle  = 90
    )
    plt.axis("equal")
    plt.tight_layout()
    plt.savefig(frames / f"{i:04}.png", dpi=150)
    plt.close()

# ---- video assembly --------------------------------------------------------
clip = ImageSequenceClip(str(frames), fps=24)       # 24 fps video
clip.write_videofile(
    f"{frames}pies.mp4",
    codec    = "libx264",
    bitrate  = "3M",
    audio    = False
)

import shutil; shutil.rmtree(frames)
```
